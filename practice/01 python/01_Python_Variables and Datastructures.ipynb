#### **기본 데이터 타입**
- 정수(integer), 실수(float), 문자열(string), 불리언(boolean)
- type 함수: 변수의 타입을 반환
- None: 변수 선언시 아무런 값을 가지지 않을때 사용

```python
a = 10
b = 11.45
print(type(a))
print(type(b))

c = None
print(c)
```

#### **비교연산자** 
- 프로그래밍에서 연산자: 비교연산자, 논리연산자
- 이들의 결과는 bool 타입

```python
a = 5
b = 4

print(a > b)  # True
print(a < b)  # False
print(a >= b) # True
print(a <= b) # False
print(a == b) # False
print(a != b) # True

c = a > b
c = True
print(type(c))
print(c)
```

#### **숫자형 데이터 타입**
- 정수 및 실수를 활용하여, 기본 사칙연산 가능

```python
a = 7
b = 4

print('a + b = ', a + b)
print('a * b = ', a * b)
print('a - b = ', a - b)
print('a / b = ', a / b)
print('a // b = ', a // b)
print('a % b = ', a % b)
print('a ** b = ', a ** b)
```

#### **연산자 우선순위**

```python
a = 5
b = 4

print(a + b * 4)
print((a + b) * 4)
```

#### **변수 값 변경**

```python
a = 9
t = a - 3
a = t
print(a)

a = 9
a = a - 3
print(a)

a = 9
a -= 3
print(a)

a = 9
a **= 3
print(a)
```
